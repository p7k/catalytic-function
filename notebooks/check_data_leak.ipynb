```python
import numpy as np
from collections import defaultdict
import pandas as pd
```

```python
path = '/scratch/spn1560'
datas = defaultdict(dict)
names = ['train', 'test']
n_splits = 2

for sidx in range(n_splits):
    for name in names:
        fn = f"{path}/sprhea_sp_folded_pt_20p_homology_threshold_80_2_splits_neg_multiple_1_seed_1234_{sidx}_split_idx_{name}.npy"
        tmp = np.load(fn)
        datas[sidx][name] = tmp
```

```python
leaks = []
for sidx in range(n_splits):
    train, test = datas[sidx]['train'], datas[sidx]['test']
    inter_rhash = set(train['feature']) & set(test['feature'])
    for row in test:
        rhash = row['feature']
        if rhash in inter_rhash:
            test_embed = row['sample_embed']
            train_rows = train[train['feature'] == rhash]
            for train_row in train_rows:
                train_embed = train_row['sample_embed']
                if (train_embed == test_embed).all():
                    leaks.append((sidx, train_row, row))
                    print(train_row, row)
```

```python
sg_path = f"{path}/sprhea_sp_folded_pt_20p_homology_threshold_80_2_splits_neg_multiple_1_seed_1234.csv"
sg = pd.read_csv(sg_path, sep='\t')
sg.head()
```

```python
gb = sg.groupby(by='split_idx')
for name, group in gb:
    pairs = list(zip(group['X1'], group['X2']))
    n_pairs = len(pairs)
    n_unq_pairs = len(set(pairs))
    if n_pairs != n_unq_pairs:
        print(n_pairs, n_unq_pairs)
```
